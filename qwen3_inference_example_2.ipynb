```python
# Установка необходимых библиотек
!pip install "unsloth[colab-new] @ git+https://github.com/unslothai/unsloth.git" -q
!pip install --no-deps "trl<0.9.0" peft accelerate bitsandbytes xformers datasets -q

# Импорт библиотек
from unsloth import FastLanguageModel
import torch

# Настройки модели
max_seq_length = 2048  # Максимальная длина контекста
dtype = None  # Автоматическое определение типа данных (Float16 для T4)
load_in_4bit = True  # Используем 4-битную квантизацию для экономии памяти

# Загрузка модели Qwen3-4B и токенизатора
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Qwen3-4B",  # Модель Qwen3-4B
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)

# Подготовка модели для инференса
FastLanguageModel.for_inference(model)

# Пример ввода в формате чата
messages = [
    {"role": "user", "content": "Привет! Расскажи, что такое искусственный интеллект?"}
]

# Применение шаблона чата
input_text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
    enable_thinking=False  # Отключаем режим "thinking" для прямого ответа
)

# Токенизация ввода
inputs = tokenizer([input_text], return_tensors="pt").to("cuda")

# Генерация ответа
outputs = model.generate(
    **inputs,
    max_new_tokens=256,  # Максимальное количество новых токенов
    temperature=0.7,     # Температура для управления креативностью
    top_p=0.9,           # Top-p sampling
    use_cache=True
)

# Декодирование ответа
response = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
print(response)
```
